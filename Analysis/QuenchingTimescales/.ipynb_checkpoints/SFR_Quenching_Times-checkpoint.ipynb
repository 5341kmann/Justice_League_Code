```python
# path = '/home/akinshol/Data/Timescales/DataFiles20190731/'
names = ['h148','h242','h229','h329']

q_thresh = 1e-11
i_thresh = 1

age = 13.800797497330507

with open('/home/akinshol/Data/Timescales/QuenchingTimescales_sSFR_S20.data','wb') as f:
    for name in names:
        print(f'Simulation {name}')
        
        z0_all = read_z0(name)
        timesteps_all = read_timesteps(name)

        for haloid in list(z0_all.haloid):
            haloids_with_stars = np.array(timesteps_all.z0haloid,dtype=int)

            if not haloid in haloids_with_stars:
                print(f'\t Halo {haloid} skipped, no stars')
                continue

            z0 = z0_all[z0_all.haloid==haloid]
            timesteps = timesteps_all[np.array(timesteps_all.z0haloid,dtype=int)==haloid]

            sfr, mstar, time = np.array(timesteps.sfr,dtype=float), np.array(timesteps.mstar,dtype=float), np.array(timesteps.time,dtype=float)
            sfr = sfr[np.argsort(time)]
            mstar = mstar[np.argsort(time)]
            time = np.sort(time)

            #s = interpolate.InterpolatedUnivariateSpline(time, mstar)

            #mstar_interpolated = s(bincenters)
            sSFR = sfr/mstar
            lbt = 13.8007 - time
            
            is_quenched = z0.sSFR.tolist()[0] < q_thresh
            
            if is_quenched:
                # then the halo is quenched, so we calculate the quenching time
                for i in range(0,len(lbt)):
                    t = np.flip(lbt)[i]
                    s = np.flip(sSFR)[i]
                    if s > 1e-11:
                        print(f'\t Halo {haloid}, Quenched (sSFR = {s}) {t} Gyr ago', end='  ')
                        tq = t
                        break
                    else:
                        continue
                        
                # lower limit on quenching time
                for i in range(1,len(timesteps)+1):
                    t = lbt[-i]
                    s = sSFR[-i]
                    if s > 2e-11:
                        tq_lower = t
                        break
                    else:
                        continue
                        
                # upper limit on quenching time
                for i in range(1,len(timesteps)+1):
                    t = lbt[-i]
                    s = sSFR[-i]
                    if s > 0:
                        tq_upper = t
                        break
                    else:
                        continue     
            else:
                print(f'\t Halo {haloid}, Unquenched at z = 0', end='  ')
                tq = None
                tq_lower = None
                tq_upper = None
                
                
            dist = np.array(timesteps.h1dist, dtype=float)
            time = age - np.array(timesteps.time, dtype=float)

            # get infall time
            try:
                ti = np.max(time[dist <= i_thresh])
                print(f'\t Halo {haloid}, infall {ti} Gyr ago')
            except ValueError:
                print(f'\t Halo {haloid}, never infell')
                ti = None
                
            try:
                ti_lower = np.max(time[dist <= 1.1])
            except ValueError:
                ti_lower = None
                
            try:
                ti_upper = np.max(time[dist <= 0.9])
            except ValueError:
                ti_upper = None
                
            z0mstar = mstar[-1]
            z0nstar = z0.n_star.tolist()[0]

            pickle.dump({
                'haloid':haloid,
                'quenched':is_quenched,
                'tquench':tq,
                'tquench_lower': tq_lower,
                'tquench_upper': tq_upper,
                'tinfall':ti,
                'tinfall_lower': ti_lower,
                'tinfall_upper': ti_upper,
                'M_star': z0mstar,
                'n_star': z0nstar,
                'sim':name                
            }, f, protocol=2)
            
```

```python

names = ['h148','h242','h229','h329']

q_thresh = 1e-11
i_thresh = 1

age = 13.800797497330507

tquenchs, tau_90s = np.array([]), np.array([])
for name in names:
    print(f'Simulation {name}')

    z0_all = read_z0(name)
    timesteps_all = read_timesteps(name)
    
    for haloid in list(np.unique(timesteps_all.z0haloid)):
        haloids_with_stars = np.array(timesteps_all.z0haloid,dtype=int)

        if not haloid in haloids_with_stars:
            print(f'\t Halo {haloid} skipped, no stars')
            continue

        z0 = z0_all[z0_all.haloid==haloid]
        timesteps = timesteps_all[np.array(timesteps_all.z0haloid,dtype=int)==haloid]

        sfr, mstar, time = np.array(timesteps.sfr,dtype=float), np.array(timesteps.mstar,dtype=float), np.array(timesteps.time,dtype=float)
        sfr = sfr[np.argsort(time)]
        mstar = mstar[np.argsort(time)]
        time = np.sort(time)

        #s = interpolate.InterpolatedUnivariateSpline(time, mstar)

        #mstar_interpolated = s(bincenters)
        sSFR = sfr/mstar
        lbt = 13.8007 - time

        is_quenched = z0.sSFR.tolist()[0] < q_thresh

        if is_quenched:
            # then the halo is quenched, so we calculate the quenching time
            for i in range(0,len(lbt)):
                t = np.flip(lbt)[i]
                s = np.flip(sSFR)[i]
                if s > 1e-11:
                    tq = t
                    break
                else:
                    continue
            
            tau_90 = z0.tquench.tolist()[0]
            try: 
                t = float(tau_90)
            except:
                continue 
                
            print(f'\t Halo {haloid}, tquench = {tq:.2f} Gyr ago, tau_90 = {tau_90:.2f} Gyr ago')
            tquenchs = np.append(tquenchs,tq)
            tau_90s = np.append(tau_90s, tau_90)

            
            

        
```
