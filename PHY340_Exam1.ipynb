```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
P = np.array([1.769, 3.551, 7.155, 16.689])*86400 # orbital period in days
a = np.array([421.6, 670.9, 1070.4, 1882.7])*1e6 # semimajor axis in 10^3 km
```

```python
plt.figure(dpi=300)
plt.scatter(np.log(a), np.log(P), label='Galilean moon data')
plt.gca().axline((np.log(a[0]),np.log(P[0])),slope=3/2, color='r', zorder=-1, label='line with slope = 1.5')
plt.legend()
plt.xlabel(r'$\ln(a)$')
plt.ylabel(r'$\ln(P)$')
plt.show()
```

```python
slope, yint = np.polyfit(np.log(a),np.log(P),deg=1)
print(f'ln(P) = {slope:.4f}*ln(a) + {yint:.4f}')
```
