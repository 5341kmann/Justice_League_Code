```python
def plot_merger_tree(z0haloid, simname):
    # tell Tangos where the simulation's tangos database is -- this h###.db file stores all the merger tree info
    import os
    os.environ['TANGOS_DB_CONNECTION'] = f'/home/akinshol/Data/Timescales/TangosData/{simname}.db'

    # import tangos and grab the simulation called 'snapshots' 
    import tangos
    s = tangos.get_simulation('snapshots')

    # specify our halo as the given haloid, from the last snapshot 
    # (i.e. if you specify z0haloid=1 at z=0, then we grab the 0th halo from the last (-1) snaphshot)
    halo = s.timesteps[-1].halos[z0haloid-1]

    # tangos has a library for generating "trees" from the progenitor data
    from tangos.relation_finding import tree 
    tree = tree.MergerTree(halo)
    tree.construct()

    # plot the tree!
    import matplotlib.pyplot as plt
    %matplotlib inline
    plt.figure(dpi=200,figsize=(10,20))
    plt.title(f'Merger Tree: {simname}-{z0haloid}')
    tree.plot()
    plt.show()
```

```python
plot_merger_tree(1, 'h329')
```

```python
plot_merger_tree(5, 'h148')
```
