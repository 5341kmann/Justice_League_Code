```python
sim = 'h148'
i = 61

z0 = read_z0(sim)
timesteps = read_timesteps(sim)

z0 = z0[z0.haloid==i]
timesteps = timesteps[timesteps.z0haloid==i]
```

```python
sfh, binEdges = z0.sfh.tolist()[0], z0.sfhbins.tolist()[0]
bincenters = 0.5*(binEdges[1:] + binEdges[:-1])
plt.plot(bincenters,sfh)
plt.xlabel('Time [Gyr]')
plt.ylabel('SFR')
plt.show()
```

```python
mstar, time = np.array(timesteps.mstar,dtype=float), np.array(timesteps.time,dtype=float)
mstar = mstar[np.argsort(time)]
time = np.sort(time)
plt.plot(time,mstar)
plt.semilogy()
plt.xlabel('Time [Gyr]')
plt.ylabel('Stellar Mass')
plt.show()
```

```python
import scipy
from scipy import interpolate

s = interpolate.InterpolatedUnivariateSpline(time, mstar)

mstar_interpolated = s(bincenters)
plt.plot(time,mstar)
plt.plot(bincenters,mstar_interpolated)
plt.semilogy()
plt.xlabel('Time [Gyr]')
plt.ylabel('Stellar Mass')
plt.show()
```

```python
sSFR = sfh/mstar_interpolated
lbt = 13.8007 - bincenters
plt.figure(dpi=100)
plt.plot(lbt, sSFR)
plt.xlabel('Lookback Time [Gyr]')
plt.ylabel(r'sSFR [yr$^{-1}$]')
plt.semilogy()
plt.gca().invert_xaxis()
plt.show()
```

```python
thresh = 1e-11
if z0.sSFR.tolist()[0] < thresh:
    # then the halo is quenched, so we calculate the quenching time
    for i in range(0, len(bincenters)):
        t = np.flip(lbt)[i]
        s = np.flip(sSFR)[i]
        print(t,s)
        if s >= thresh:
            print(f'Quenched (sSFR = {s}) {t} Gyr ago')
            break
        else:
            continue
else:
    print('Unquenched')
```

```python
path = '/home/akinshol/Data/Timescales/DataFiles20190731/'
names = ['h148','h242','h229','h329']

q_thresh = 1e-11
i_thresh = 1

age = 13.800797497330507

with open('/home/akinshol/Data/Timescales/QuenchingTimescales_sSFR_F19.data','wb') as f:
    for name in names:
        print(f'Simulation {name}')
        
        z0_all = read_z0(name)
        timesteps_all = read_timesteps(name)

        for haloid in list(z0_all.haloid):
            haloids_with_stars = np.array(timesteps_all.z0haloid,dtype=int)

            if not haloid in haloids_with_stars:
                print(f'\t Halo {haloid} skipped, no stars')
                continue

            z0 = z0_all[z0_all.haloid==haloid]
            timesteps = timesteps_all[np.array(timesteps_all.z0haloid,dtype=int)==haloid]

            sfh, binEdges = z0.sfh.tolist()[0], z0.sfhbins.tolist()[0]
            bincenters = 0.5*(binEdges[1:] + binEdges[:-1])

            mstar, time = np.array(timesteps.mstar,dtype=float), np.array(timesteps.time,dtype=float)
            mstar = mstar[np.argsort(time)]
            time = np.sort(time)

            s = interpolate.InterpolatedUnivariateSpline(time, mstar)

            mstar_interpolated = s(bincenters)
            sSFR = sfh/mstar_interpolated
            lbt = age - bincenters
            
            is_quenched = z0.sSFR.tolist()[0] < q_thresh
            
            if is_quenched:
                # then the halo is quenched, so we calculate the quenching time
                for i in range(0,len(lbt)):
                    t = np.flip(lbt)[i]
                    s = np.flip(sSFR)[i]
                    if s > 1e-11:
                        print(f'\t Halo {haloid}, Quenched (sSFR = {s}) {t} Gyr ago', end='  ')
                        tq = t
                        break
                    else:
                        continue
                        
                # lower limit on quenching time
                for i in range(1,len(timesteps)+1):
                    t = lbt[-i]
                    s = sSFR[-i]
                    if s > 2e-11:
                        tq_lower = t
                        break
                    else:
                        continue
                        
                # upper limit on quenching time
                for i in range(1,len(timesteps)+1):
                    t = lbt[-i]
                    s = sSFR[-i]
                    if s > 0:
                        tq_upper = t
                        break
                    else:
                        continue     
            else:
                print(f'\t Halo {haloid}, Unquenched at z = 0', end='  ')
                tq = None
                tq_lower = None
                tq_upper = None
                
                
            dist = np.array(timesteps.h1dist, dtype=float)
            time = age - np.array(timesteps.time, dtype=float)

            # get infall time
            try:
                ti = np.max(time[dist <= i_thresh])
                print(f'\t Halo {haloid}, infall {ti} Gyr ago')
            except ValueError:
                print(f'\t Halo {haloid}, never infell')
                ti = None
                
            try:
                ti_lower = np.max(time[dist <= 1.1])
            except ValueError:
                ti_lower = None
                
            try:
                ti_upper = np.max(time[dist <= 0.9])
            except ValueError:
                ti_upper = None
                
            z0mstar = mstar[-1]
            z0nstar = z0.n_star.tolist()[0]

            pickle.dump({
                'haloid':haloid,
                'quenched':is_quenched,
                'tquench':tq,
                'tquench_lower': tq_lower,
                'tquench_upper': tq_upper,
                'tinfall':ti,
                'tinfall_lower': ti_lower,
                'tinfall_upper': ti_upper,
                'M_star': z0mstar,
                'n_star': z0nstar,
                'sim':name                
            }, f, protocol=2)
            
```

```python
def read_file(path, halo_num):
    '''Function to read in the timestep bulk-processing datafile (from /home/akinhol/Data/Timescales/DataFiles/{name}.data)'''
    data = []
    with open(path,'rb') as f:
        while True:
            try: 
                data.append(pickle.load(f))
            except EOFError:
                break
    data = pd.DataFrame(data)
    
    if not halo_num=='all':
        data = data[data.z0haloid == halo_num]

    return data

current_data = read_file('/home/akinshol/Data/Timescales/DataFiles20190731/h148.data', 10)
time = 13.8 - np.array(current_data.time, dtype=float)

np.max(time[np.array(current_data.h1dist) <= 1])
```
