```python
import pynbody
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.axes_grid1 import make_axes_locatable
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
from matplotlib.colors import LogNorm
```

```python
s = pynbody.load('/home/christenc/Data/Sims/h148.cosmo50PLK.3072g/h148.cosmo50PLK.3072g3HbwK1BH/snapshots/h148.cosmo50PLK.3072g3HbwK1BH.004096')
s.physical_units()
h = s.halos()
print('Centering...')
pynbody.analysis.halo.center(h[1])
```

```python
rvir = h[1].properties['Rvir']/s.properties['h']
print(rvir)
```

```python
pynbody.analysis.angmom.sideon(h[1])
```

```python
width = 1000
res = 500


fig, ax = plt.subplots(1,3,figsize=(11,4),dpi=300)

print('Making Images...')

# dm density
im = pynbody.plot.sph.image(s.dm, width =f'{str(width)} kpc', cmap ='gray', resolution=res, av_z = 'rho', noplot=True)
im = ax[0].imshow(im, norm=LogNorm(), cmap='gray')

divider = make_axes_locatable(ax[0])
cax = divider.new_vertical(size="5%", pad=0.1, pack_start=True)
fig.add_axes(cax)
cbar = fig.colorbar(im, cax=cax, orientation="horizontal")
cbar.set_label('Dark Matter Density [$\mathrm{M}_{\odot}\ \mathrm{kpc}^{-3}$]')

circ = mpl.patches.Circle((res/2,res/2),radius=res*rvir/width,ec='r',fc='none')
ax[0].add_patch(circ)

# # gas density
# im = pynbody.plot.sph.image(s.g, width = '300 kpc', noplot=True, av_z = 'rho')
# im = ax[1].imshow(im, norm=LogNorm(), cmap='viridis')

# divider = make_axes_locatable(ax[1])
# cax = divider.new_vertical(size="5%", pad=0.1, pack_start=True)
# fig.add_axes(cax)
# cbar = fig.colorbar(im, cax=cax, orientation="horizontal")
# cbar.set_label('Gas Density [$\mathrm{M}_{\odot}\ \mathrm{kpc}^{-3}$]')



# # stellar density
# im = pynbody.plot.sph.image(s.s, width = '300 kpc', noplot=True, av_z = 'rho')
# im = ax[2].imshow(im, norm=LogNorm(), cmap='cubehelix', vmin=1e2)#, vmin=5e3, vmax=1e9)

# divider = make_axes_locatable(ax[2])
# cax = divider.new_vertical(size="5%", pad=0.1, pack_start=True)
# fig.add_axes(cax)
# cbar = fig.colorbar(im, cax=cax, orientation="horizontal")
# cbar.set_label('Stellar Density [$\mathrm{M}_{\odot}\ \mathrm{kpc}^{-3}$]')

# scalebar = AnchoredSizeBar(ax[2].transData,167, '100 kpc', 'upper left', pad=0.5,color='white',frameon=False,size_vertical=1)
# ax[2].add_artist(scalebar)


for i in [0,1,2]:
    ax[i].tick_params(labelcolor='none', top='off', bottom='off', left='off', right='off')
    ax[i].set_xticks([])
    ax[i].set_yticks([])
    



plt.subplots_adjust(wspace=0.05)
plt.show()
```
