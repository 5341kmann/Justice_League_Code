# Here's how we calculated the (temporary) estimate of the background density

First we load in the simulation (in this case, h329, snapshot 4096).

```python
s = pynbody.load('/home/christenc/Data/Sims/h329.cosmo50PLK.3072g/h329.cosmo50PLK.3072gst5HbwK1BH/snapshots/h329.cosmo50PLK.3072gst5HbwK1BH.004096')
s.physical_units()
h = s.halos()

h1rvir_old = h[1].properties['Rvir']
```

We then center the halo and make a spherical density profile for all matter in the simulation.

```python
pynbody.analysis.halo.center(h[1])
```

```python
p = pynbody.analysis.profile.Profile(s, ndim = 3, min=0.001,max=6*h1rvir_old) 
```

We use this critical density function from pynbody to replicate the way AHF defines the virial radius as $R_{200,c}$. Then we can use that value to esimate the radius $R_{200,b}$. 

```python
critical_density = pynbody.analysis.cosmology.rho_crit(s)
```

```python
critical_density
```

Here we plot the density profile

```python
plt.figure(dpi=150)
plt.plot(p['rbins'],p['density_enc'])
plt.semilogy()
# plt.plot([np.min(p['rbins']), np.max(p['rbins'])], [200*background_density, 200*background_density], 'r--')
plt.plot([np.min(p['rbins']), np.max(p['rbins'])], [200*critical_density, 200*critical_density], 'g--')
plt.show()

print('Estimate virial radius using 200 times critical density:',np.min(p['rbins'][p['density_enc'] < 200*critical_density]), 'AHF virial radius in physical units:',h1rvir_old/s.properties['h'] * s.properties['a'])

print('Enclosed density at 1.92 times R_200,c:',np.max(p['density_enc'][p['rbins'] > 1.92*h1rvir_old/s.properties['h'] * s.properties['a']]))
```
