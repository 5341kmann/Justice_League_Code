```python
data = read_file(h148,'h148')

i=45

sfh = data['sfh'].tolist()[i]
bins = data['sfhbins'].tolist()[i]

### quenching time 
sfhmstar = np.sum(sfh)
# create cumulative sfh 
bincenters = 0.5*(bins[1:]+bins[:-1])
c_sfh = np.empty(shape=sfh.shape)
for i in range(len(bincenters)):
    c_sfh[i] = np.sum(sfh[:i+1])/sfhmstar


    
time = np.min(bincenters[c_sfh > 0.9])

age = 13.800797497330507
tquench = age - time

plt.plot(age - bincenters,c_sfh)
plt.gca().invert_xaxis()
plt.show()

print(tquench)
```

### Quenching Time

First need to figure out the time of quenching

```python
data = read_all()

data = data[data.n_star > 100]
q = quenched(data)

tquench = data['tquench']

fig,ax = plt.subplots(1,1,dpi=300,figsize=(4.5,2.8))
ax.scatter(tquench[q],data.M_star[q],s=20,linewidths=0.7,edgecolor='r',facecolor='none',label='quenched')
ax.scatter(tquench[~q],data.M_star[~q],s=20,linewidths=0.7,edgecolor='b',facecolor='none',label='star-forming')
ax.legend()
ax.invert_xaxis()
ax.semilogy()
ax.set_xlabel('Lookback "Quenching" Time ($t_{90}$) [Gyr]')
ax.set_ylabel('Stellar Mass ($M_*$) [$\mathrm{M}_{\odot}$]')
plt.show()

#     bincenters = age - bincenters 
#     fig,ax = plt.subplots(1,1,figsize=(4,4),dpi=150)
#     ax.scatter(bincenters,c_sfh,s=4)
#     ax.plot([0,13.8],[.9,.9],linestyle=':',color='k',alpha=0.3)
#     ax.plot([t90,t90],[-1,2],linestyle=':',color='k',alpha=0.3)
#     ax.set_xlim(np.min(bincenters),np.max(bincenters))
#     ax.set_ylim(-0.1,1.1)
#     ax.invert_xaxis()
#     ax.set_xlabel('Lookback time (Gyr)')
#     ax.set_ylabel('Cumulative SFH')
#     plt.show()
```
